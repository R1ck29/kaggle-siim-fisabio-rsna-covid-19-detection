```python
import pandas as pd
from sklearn.model_selection import GroupKFold
```

```python
df = pd.read_csv('../input/train_image_level.csv')
df['class_id'] = 0
for i in range(df.shape[0]):
    if df.loc[i,'label'] == 'none 1 0 0 1 1':
        df.loc[i,'class_id'] = 1
```

```python
df
```

```python
label_cols = df.columns[4]
```

```python
label_cols
```

```python
gkf  = GroupKFold(n_splits = 5)
df['fold'] = -1
for fold, (train_idx, val_idx) in enumerate(gkf.split(df, groups = df.StudyInstanceUID.tolist())):
    df.loc[val_idx, 'fold'] = fold
```

```python
df
```

```python
for fold in range(5):
    fold_df = df[df['fold']==fold]
    print(fold_df.class_id.value_counts())
```

```python
df.to_csv('../input/train_image_level_2class.csv',index=False)
```
